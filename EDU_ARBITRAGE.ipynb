# Количественный арбитраж (Quantitative arbitrage)
это автоматизированная торговля на основе количественных моделей, направленная на извлечение прибыли из краткосрочных дисбалансов в ценах между связанными активами или инструментами

# Одна из стратегий Парный арбитраж (Pair-Trading)

нам необходимо выявить два (или больше) коррелированных актива, у которых есть временное расхождение в цене, и зарабатывать на возвращении к среднему. Методы оценки таких стратегий включают статистические подходы, проверяющие наличие устойчивой зависимости и возможность извлечения прибыли

# Методы оценки

## 1. Корреляционный анализ (Pearson correlation coefficient)

* Проверяет линейную связь между двумя временными рядами
* Используется как первичная фильтрация: отбираются пары с высокой исторической корреляцией (например, > 0.8)

#### Коэффициент корреляции Пирсона (Pearson correlation coefficient) — это статистическая мера линейной зависимости между двумя переменными. Он обозначается символом r и принимает значения от -1 до 1, где:

r = 1 — идеальная положительная линейная корреляция
r = 0 — отсутствие линейной корреляции
r = -1 — идеальная отрицательная линейная корреляция

$$r = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2 \sum_{i=1}^{n} (y_i - \bar{y})^2}}$$

Где:
- $r$ — коэффициент корреляции Пирсона
- $x_i, y_i$ — значения переменных X и Y
- $\bar{x}$ — среднее арифметическое переменной X
- $\bar{y}$ — среднее арифметическое переменной Y
- $n$ — размер выборки (количество пар значений)

Альтернативная формула для вычислений:

$$r = \frac{n\sum_{i=1}^{n} x_i y_i - \sum_{i=1}^{n} x_i \sum_{i=1}^{n} y_i}{\sqrt{n\sum_{i=1}^{n} x_i^2 - (\sum_{i=1}^{n} x_i)^2} \sqrt{n\sum_{i=1}^{n} y_i^2 - (\sum_{i=1}^{n} y_i)^2}}$$

## 2. Коинтеграция (Cointegration)
это статистическое свойство временных рядов, которое описывает долгосрочное равновесное соотношение между двумя или более нестационарными временными рядами. Если простыми словами, это означает, что даже если отдельные временные ряды нестабильны и имеют тренды (то есть являются нестационарными), может существовать такая их линейная комбинация, которая стационарна

### Тест Энгла-Грейнджера на коинтеграцию

#### Шаг 1: Оценка долгосрочной взаимосвязи
Оцените регрессионное уравнение долгосрочной связи методом наименьших квадратов:

$$y_t = \beta_0 + \beta_1 x_t + u_t$$, где:
- $y_t$ и $x_t$ — исследуемые временные ряды
- $\beta_0$ и $\beta_1$ — коэффициенты регрессии
- $u_t$ — остатки регрессии

#### Шаг 2: Проверка стационарности остатков
Проведите расширенный тест Дики-Фуллера (ADF) для остатков $\hat{u}_t$:

$$\Delta \hat{u}_t = \alpha \hat{u}_{t-1} + \sum_{i=1}^{p} \gamma_i \Delta \hat{u}_{t-i} + \varepsilon_t$$, где:
- $\Delta \hat{u}_t$ — первые разности остатков
- $\hat{u}_{t-1}$ — лагированные значения остатков
- $\alpha$ — коэффициент, проверяемый на значимость
- $\gamma_i$ — коэффициенты при лагированных разностях
- $\varepsilon_t$ — случайная ошибка
- $p$ — число лагов, включенных в модель

Тестовая статистика:
$$\tau = \frac{\hat{\alpha}}{SE(\hat{\alpha})}$$, где $SE(\hat{\alpha})$ — стандартная ошибка оценки коэффициента $\alpha$

#### Шаг 3: P-value (p-значение) — это статистический показатель, используемый для оценки значимости результатов статистического теста
P-value — это вероятность получения наблюдаемых (или более экстремальных) результатов при условии, что нулевая гипотеза верна

Формально записывается как:
$$p = P(T \geq t_{obs} | H_0)$$, где:
* $T$ — случайная величина, соответствующая тестовой статистике
* $t_{obs}$ — наблюдаемое значение тестовой статистики
* $H_0$ — нулевая гипотеза

Чем меньше p-value, тем выше вероятность коинтеграции между активами

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from binance.client import Client
from datetime import datetime
import seaborn as sns
from statsmodels.tsa.stattools import coint
import itertools
import warnings
warnings.filterwarnings('ignore')

# Настройка клиента Binance с API ключами
api_key = ""
api_secret = ""
client = Client(api_key, api_secret)

# Список активов
assets = [
    "BTCUSDT", "BNBUSDT", "WIFUSDT", "1000PEPEUSDT", "ENAUSDT", 
    "ARBUSDT", "OPUSDT", "LTCUSDT", "WLDUSDT", "DOTUSDT", 
    "ONDOUSDT", "SOLUSDT", "SUIUSDT", "XRPUSDT", "DOGEUSDT", 
    "TRUMPUSDT", "ADAUSDT", "TRXUSDT", "LINKUSDT", "AVAXUSDT"
]

# Установка временного диапазона
start_date = "2025-01-01"
end_date = "2025-05-01"

# Преобразование дат в миллисекунды для API Binance
start_ms = int(datetime.strptime(start_date, "%Y-%m-%d").timestamp() * 1000)
end_ms = int(datetime.strptime(end_date, "%Y-%m-%d").timestamp() * 1000)

# Функция для получения данных
def get_perpetual_data(symbol, interval='1h', start=start_ms, end=end_ms):
    print(f"Загрузка данных для {symbol}...")
    
    # Получение данных фьючерсов (perpetual)
    klines = client.futures_klines(symbol=symbol, interval=interval, startTime=start, endTime=end, limit=1000)
    
    # Обработка данных свечей
    data = []
    for k in klines:
        data.append([
            datetime.fromtimestamp(k[0]/1000),  # время открытия
            float(k[4])  # цена закрытия
        ])
    
    # Создание DataFrame
    df = pd.DataFrame(data, columns=['timestamp', symbol])
    df.set_index('timestamp', inplace=True)
    
    return df

# Загрузка данных для всех активов
dfs = []
for asset in assets:
    try:
        df = get_perpetual_data(asset)
        if not df.empty:
            dfs.append(df)
        else:
            print(f"Не удалось получить данные для {asset}")
    except Exception as e:
        print(f"Ошибка при получении данных для {asset}: {e}")

# Объединение всех данных в один DataFrame
if dfs:
    data = pd.concat(dfs, axis=1)
    data.fillna(method='ffill', inplace=True)  # Заполнение пропущенных значений
    
    # Сохранение данных в CSV для будущего использования
    data.to_csv('perpetual_data.csv')
    print("Данные успешно сохранены в файл perpetual_data.csv")
else:
    print("Не удалось получить данные. Проверьте подключение или API ключи.")
    exit()

# Расчет корреляции
correlation_matrix = data.pct_change().corr()

# Построение тепловой карты корреляции
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Корреляция между активами (5m таймфрейм)', fontsize=16)
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
print("Тепловая карта корреляции сохранена в файл correlation_heatmap.png")

# Создание матрицы коинтеграции (p-value)
def create_cointegration_matrix(data):
    assets = data.columns
    n = len(assets)
    coint_matrix = pd.DataFrame(np.ones((n, n)), index=assets, columns=assets)
    
    for i, asset1 in enumerate(assets):
        for j, asset2 in enumerate(assets):
            if i >= j:  # Только верхняя треугольная часть
                continue
                
            series1 = data[asset1].values
            series2 = data[asset2].values
            
            # Проверка на NaN
            if np.isnan(series1).any() or np.isnan(series2).any():
                coint_matrix.loc[asset1, asset2] = np.nan
                continue
                
            # Расчет коинтеграции (p-value)
            _, pvalue, _ = coint(series1, series2)
            coint_matrix.loc[asset1, asset2] = pvalue
            coint_matrix.loc[asset2, asset1] = pvalue  # Симметричность
    
    # Диагональ
    np.fill_diagonal(coint_matrix.values, 0)
    
    return coint_matrix

# Создание матрицы коинтеграции
coint_matrix = create_cointegration_matrix(data)

# Построение тепловой карты коинтеграции
plt.figure(figsize=(20, 16))
# Чем меньше p-value, тем больше вероятность коинтеграции, 
# поэтому используем обратную цветовую шкалу (меньшие значения - более яркие)
mask = np.triu(np.ones_like(coint_matrix, dtype=bool), k=1)  # Только верхний треугольник
sns.heatmap(coint_matrix, annot=True, mask=~mask, cmap='YlGnBu_r', vmin=0, vmax=0.05, linewidths=0.5, 
           annot_kws={"size": 8})
plt.title('P-value коинтеграции между активами (5m таймфрейм)\nЧем ниже p-value, тем выше вероятность коинтеграции', fontsize=16)
plt.tight_layout()
plt.savefig('cointegration_heatmap.png')
print("Тепловая карта коинтеграции сохранена в файл cointegration_heatmap.png")

# Функция для проверки коинтеграции (полный список пар)
def check_cointegration(data):
    pairs = []
    
    # Создание всех возможных пар активов
    combinations = list(itertools.combinations(data.columns, 2))
    
    for (asset1, asset2) in combinations:
        series1 = data[asset1].values
        series2 = data[asset2].values
        
        # Проверка на NaN
        if np.isnan(series1).any() or np.isnan(series2).any():
            continue
            
        # Расчет коинтеграции
        score, pvalue, _ = coint(series1, series2)
        
        # Корреляция между активами
        corr = correlation_matrix.loc[asset1, asset2]
        
        # Сохранение результатов
        pairs.append((asset1, asset2, score, pvalue, corr))
    
    # Создание DataFrame с результатами
    pairs_df = pd.DataFrame(pairs, columns=['Asset1', 'Asset2', 'Score', 'PValue', 'Correlation_Value'])
    
    # Сортировка по p-value (чем меньше, тем выше вероятность коинтеграции)
    pairs_df = pairs_df.sort_values(by='PValue')
    
    return pairs_df

# Проверка коинтеграции
cointegration_results = check_cointegration(data)

# Сохранение результатов в CSV
cointegration_results.to_csv('cointegration_results.csv', index=False)
print("Результаты коинтеграции сохранены в файл cointegration_results.csv")

# Вывод 10 лучших пар с наивысшей коинтеграцией
print("\nТоп-10 пар с наивысшей коинтеграцией (низкий p-value):")
top_coint_pairs = cointegration_results.head(10)
print(top_coint_pairs)

# Вывод 10 лучших пар с наивысшей корреляцией
corr_pairs = []
for i, asset1 in enumerate(correlation_matrix.columns):
    for asset2 in correlation_matrix.columns[i+1:]:
        corr_pairs.append((asset1, asset2, correlation_matrix.loc[asset1, asset2]))

corr_df = pd.DataFrame(corr_pairs, columns=['Asset1', 'Asset2', 'Correlation_Value'])
corr_df = corr_df.sort_values(by='Correlation_Value', ascending=False)

print("\nТоп-10 пар с наивысшей корреляцией:")
top_corr_pairs = corr_df.head(10)
print(top_corr_pairs)

# Определение лучших пар по комбинации корреляции и коинтеграции
try:
    # Объединение данных о корреляции и коинтеграции
    combined_results = pd.merge(
        corr_df, 
        cointegration_results[['Asset1', 'Asset2', 'PValue', 'Score']], 
        on=['Asset1', 'Asset2']
    )
    
    # Подготовка ранжирования
    combined_results['AbsCorrelation'] = combined_results['Correlation_Value'].abs()
    combined_results['RankCorr'] = combined_results['AbsCorrelation'].rank(ascending=False)
    combined_results['RankCoint'] = combined_results['PValue'].rank()
    combined_results['CombinedRank'] = combined_results['RankCorr'] + combined_results['RankCoint']
    
    # Сортировка по комбинированному рангу
    combined_results = combined_results.sort_values(by='CombinedRank')
    
    print("\nТоп-10 пар по комбинации корреляции и коинтеграции:")
    top_combined_pairs = combined_results[['Asset1', 'Asset2', 'Correlation_Value', 'PValue', 'CombinedRank']].head(10)
    print(top_combined_pairs)
    
    # Сохранение результатов в CSV
    combined_results.to_csv('combined_ranking.csv', index=False)
    print("Комбинированный рейтинг пар сохранен в файл combined_ranking.csv")
    
    # Визуализация лучших пар
    def plot_pair(data, pair_info):
        asset1, asset2 = pair_info['Asset1'], pair_info['Asset2']
        corr = pair_info['Correlation_Value']
        pvalue = pair_info['PValue']
        
        plt.figure(figsize=(14, 7))
        
        # Нормализация для сравнения
        norm1 = (data[asset1] - data[asset1].min()) / (data[asset1].max() - data[asset1].min())
        norm2 = (data[asset2] - data[asset2].min()) / (data[asset2].max() - data[asset2].min())
        
        plt.plot(norm1, label=asset1)
        plt.plot(norm2, label=asset2)
        
        plt.title(f'Сравнение {asset1} и {asset2}\nКорреляция: {corr:.4f}, P-value коинтеграции: {pvalue:.4f}')
        plt.legend()
        plt.grid(True)
        plt.savefig(f'pair_{asset1}_{asset2}.png')
        plt.close()
    
    # Построение графиков для топ-5 пар
    for i, row in combined_results.head(5).iterrows():
        plot_pair(data, row)
        print(f"График для пары {row['Asset1']} и {row['Asset2']} сохранен")
    
    # АНАЛИЗ И ВЫВОДЫ
    # Определим пары с высокой корреляцией и хорошей коинтеграцией (потенциальные пары для торговли)
    strong_pairs = combined_results[
        (combined_results['AbsCorrelation'] > 0.8) & 
        (combined_results['PValue'] < 0.05)
    ].sort_values('CombinedRank')
    
    # Создаем выводы на основании анализа
    print("\n\n=====================================================")
    print("АНАЛИЗ И ВЫВОДЫ ПО ИССЛЕДОВАНИЮ КРИПТОАКТИВОВ")
    print("=====================================================")
    
    # 1. Общая картина корреляций
    print("\n1. АНАЛИЗ КОРРЕЛЯЦИЙ:")
    print(f"- Всего было проанализировано {len(assets)} активов на предмет корреляции.")
    
    # Подсчет средней и медианной корреляции
    avg_corr = corr_df['Correlation_Value'].mean()
    median_corr = corr_df['Correlation_Value'].median()
    max_corr = corr_df['Correlation_Value'].max()
    min_corr = corr_df['Correlation_Value'].min()
    
    print(f"- Средняя корреляция между всеми активами: {avg_corr:.4f}")
    print(f"- Медианная корреляция: {median_corr:.4f}")
    print(f"- Максимальная положительная корреляция: {max_corr:.4f} (между {top_corr_pairs.iloc[0]['Asset1']} и {top_corr_pairs.iloc[0]['Asset2']})")
    
    # Определяем наиболее коррелированные группы активов
    high_corr_groups = {}
    for asset in assets:
        high_corr_assets = correlation_matrix[correlation_matrix[asset] > 0.8].index.tolist()
        high_corr_assets = [a for a in high_corr_assets if a != asset]
        if high_corr_assets:
            high_corr_groups[asset] = high_corr_assets
    
    # Выводим группы с высокой корреляцией (если есть)
    if high_corr_groups:
        print("\n- Группы активов с высокой корреляцией (> 0.8):")
        for asset, correlated_assets in high_corr_groups.items():
            if correlated_assets:
                print(f"  * {asset} сильно коррелирует с: {', '.join(correlated_assets)}")
    
    # 2. Анализ коинтеграции
    print("\n2. АНАЛИЗ КОИНТЕГРАЦИИ:")
    print(f"- Пар с высокой вероятностью коинтеграции (p-value < 0.05): {len(cointegration_results[cointegration_results['PValue'] < 0.05])}")
    print(f"- Пар с очень высокой вероятностью коинтеграции (p-value < 0.01): {len(cointegration_results[cointegration_results['PValue'] < 0.01])}")
    
    if not top_coint_pairs.empty:
        best_coint_pair = top_coint_pairs.iloc[0]
        print(f"- Наилучшая коинтегрированная пара: {best_coint_pair['Asset1']} и {best_coint_pair['Asset2']} (p-value: {best_coint_pair['PValue']:.6f})")
    
    # 3. Комбинированный анализ (корреляция + коинтеграция)
    print("\n3. КОМБИНИРОВАННЫЙ АНАЛИЗ (КОРРЕЛЯЦИЯ + КОИНТЕГРАЦИЯ):")
    print(f"- Найдено {len(strong_pairs)} пар с высокой корреляцией (>0.8) и значительной коинтеграцией (p<0.05)")
    
    if not strong_pairs.empty:
        print("\n- Лучшие пары для потенциального парного трейдинга:")
        for i, row in strong_pairs.head(5).iterrows():
            print(f"  * {row['Asset1']} и {row['Asset2']} - Корреляция: {row['Correlation_Value']:.4f}, p-value коинтеграции: {row['PValue']:.6f}")
    
    # 4. Заключение и рекомендации
    print("\n4. ЗАКЛЮЧЕНИЕ И РЕКОМЕНДАЦИИ:")
    print("- Для парного трейдинга рекомендуется выбирать пары с высокой корреляцией (>0.8) и низким p-value коинтеграции (<0.05).")
    
    if not strong_pairs.empty:
        best_pair = strong_pairs.iloc[0]
        print(f"- Наиболее перспективная пара для трейдинга: {best_pair['Asset1']} и {best_pair['Asset2']}")
        print(f"  Корреляция: {best_pair['Correlation_Value']:.4f}, p-value коинтеграции: {best_pair['PValue']:.6f}")
        
    # Анализ специфических активов
    btc_pairs = combined_results[combined_results['Asset1'] == 'BTCUSDT'].sort_values('CombinedRank')
    if not btc_pairs.empty:
        print("\n- Лучшие пары с Bitcoin (BTCUSDT):")
        for i, row in btc_pairs.head(3).iterrows():
            print(f"  * BTCUSDT и {row['Asset2']} - Корреляция: {row['Correlation_Value']:.4f}, p-value: {row['PValue']:.6f}")
    
    # Сохранение выводов в файл
    with open('analysis_conclusions.txt', 'w', encoding='utf-8') as f:
        f.write("АНАЛИЗ И ВЫВОДЫ ПО ИССЛЕДОВАНИЮ КРИПТОАКТИВОВ\n")
        f.write("=====================================================\n\n")
        f.write("1. АНАЛИЗ КОРРЕЛЯЦИЙ:\n")
        f.write(f"- Всего было проанализировано {len(assets)} активов на предмет корреляции.\n")
        f.write(f"- Средняя корреляция между всеми активами: {avg_corr:.4f}\n")
        f.write(f"- Медианная корреляция: {median_corr:.4f}\n")
        f.write(f"- Максимальная положительная корреляция: {max_corr:.4f} (между {top_corr_pairs.iloc[0]['Asset1']} и {top_corr_pairs.iloc[0]['Asset2']})\n\n")
        
        if high_corr_groups:
            f.write("- Группы активов с высокой корреляцией (> 0.8):\n")
            for asset, correlated_assets in high_corr_groups.items():
                if correlated_assets:
                    f.write(f"  * {asset} сильно коррелирует с: {', '.join(correlated_assets)}\n")
        
        f.write("\n2. АНАЛИЗ КОИНТЕГРАЦИИ:\n")
        f.write(f"- Пар с высокой вероятностью коинтеграции (p-value < 0.05): {len(cointegration_results[cointegration_results['PValue'] < 0.05])}\n")
        f.write(f"- Пар с очень высокой вероятностью коинтеграции (p-value < 0.01): {len(cointegration_results[cointegration_results['PValue'] < 0.01])}\n")
        
        if not top_coint_pairs.empty:
            best_coint_pair = top_coint_pairs.iloc[0]
            f.write(f"- Наилучшая коинтегрированная пара: {best_coint_pair['Asset1']} и {best_coint_pair['Asset2']} (p-value: {best_coint_pair['PValue']:.6f})\n")
        
        f.write("\n3. КОМБИНИРОВАННЫЙ АНАЛИЗ (КОРРЕЛЯЦИЯ + КОИНТЕГРАЦИЯ):\n")
        f.write(f"- Найдено {len(strong_pairs)} пар с высокой корреляцией (>0.8) и значительной коинтеграцией (p<0.05)\n\n")
        
        if not strong_pairs.empty:
            f.write("- Лучшие пары для потенциального парного трейдинга:\n")
            for i, row in strong_pairs.head(5).iterrows():
                f.write(f"  * {row['Asset1']} и {row['Asset2']} - Корреляция: {row['Correlation_Value']:.4f}, p-value коинтеграции: {row['PValue']:.6f}\n")
        
        f.write("\n4. ЗАКЛЮЧЕНИЕ И РЕКОМЕНДАЦИИ:\n")
        f.write("- Для парного трейдинга рекомендуется выбирать пары с высокой корреляцией (>0.8) и низким p-value коинтеграции (<0.05).\n")
        
        if not strong_pairs.empty:
            best_pair = strong_pairs.iloc[0]
            f.write(f"- Наиболее перспективная пара для трейдинга: {best_pair['Asset1']} и {best_pair['Asset2']}\n")
            f.write(f"  Корреляция: {best_pair['Correlation_Value']:.4f}, p-value коинтеграции: {best_pair['PValue']:.6f}\n")
        
        if not btc_pairs.empty:
            f.write("\n- Лучшие пары с Bitcoin (BTCUSDT):\n")
            for i, row in btc_pairs.head(3).iterrows():
                f.write(f"  * BTCUSDT и {row['Asset2']} - Корреляция: {row['Correlation_Value']:.4f}, p-value: {row['PValue']:.6f}\n")
    
    print("\nПолный анализ и выводы сохранены в файл 'analysis_conclusions.txt'")
    
except Exception as e:
    print(f"Произошла ошибка при объединении данных: {e}")
    print("Трассировка ошибки:")
    import traceback
    traceback.print_exc()
    
    # Альтернативный способ определения лучших пар
    print("\nИспользую альтернативный метод для поиска лучших пар:")
    
    # Создаем отдельный DataFrame для вывода лучших пар по корреляции
    print("\nТоп-10 пар по корреляции:")
    print(corr_df.head(10))
    
    # Создаем отдельный DataFrame для вывода лучших пар по коинтеграции
    print("\nТоп-10 пар по коинтеграции (низкий p-value):")
    print(cointegration_results.sort_values(by='PValue').head(10))

print("\nАнализ завершен.")
```
