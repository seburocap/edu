## 1.1 Технологический риск инфраструктуры
Риск отказа или некорректной работы торговой инфраструктуры, включая сбои в работе серверов, проблемы с сетевой связностью, задержки в исполнении ордеров, а также уязвимости в программном обеспечении алгоритмов

### Хеджирование
* Развертывание распределенной инфраструктуры с географической диверсификацией серверов, внедрение систем автоматического переключения между резервными узлами, использование множественных провайдеров связи, регулярное тестирование систем восстановления после сбоев

### 1.2 Регуляторный риск
Риск изменения регуляторной среды, включая запреты на торговлю определенными активами, ужесточение требований к лицензированию, изменения налогового законодательства в различных юрисдикциях

### Хеджирование
* Юридическая структуризация фонда в благоприятных юрисдикциях, постоянный мониторинг регуляторных изменений, поддержание соответствия KYC/AML требованиям, диверсификация юридического присутствия, создание резервов под потенциальные налоговые обязательства

### 1.3 Риск кибербезопасности
Угроза несанкционированного доступа к торговым системам, кражи приватных ключей, фишинговых атак на сотрудников, а также DDoS атак на торговую инфраструктуру

### Хеджирование
* Внедрение многофакторной аутентификации для как входа в биржу, так и хранения ключей например к холодному кошельку где хранятся средства, регулярные пенетрационные тесты (не сайта, а всех кто задействован в переводе средств на и из биржи), обучение команды основам кибербезопасности, страхование киберрисков

### 1.4 Риск операционной масштабируемости
Неспособность инфраструктуры и операционных процессов эффективно масштабироваться при росте AUM (assets under management), что приводит к снижению доходности и увеличению операционных рисков

### Хеджирование
* Инвестиции в масштабируемую инфраструктуру, автоматизация процессов, установление жестких лимитов на максимальный AUM для каждой стратегии в $ или %, расписать план аллокаций для каждой стратегии в будущем

### 1.5 Технологический риск инфраструктуры
Сбои в работе торговых систем биржи, включая недоступность API, задержки в обработке ордеров, некорректное исполнение или отмена транзакций

### Хеджирование
* Резервирование API-подключений и использование нескольких endpoint'ов
* Локальное кэширование состояния ордеров
* Разработка алгоритмов восстановления после сбоев
* Фокус отдельного специалиста на таких проблемах

### 1.6 Риск кастодиальной безопасности (Custodial Security Risk)
Компрометация приватных ключей, компроментация аккаунта биржи и тд

### Хеджирование

### 2.1 Риск концентрации компетенций (Key Person Risk) 
* Критическая зависимость торговых стратегий и операционных процессов от узкого круга специалистов

### Хеджирование
* Внедрение системы парного программирования и обязательного код-ревью
* Документирование всех критических процессов и алгоритмов
* Ротация специалистов между проектами (повышает эффективность разработки стратегии)

#### Индекс Херфиндаля-Хиршмана (HHI)
HHI = Σ(si)² × 10000, где:
* si - доля критических задач/компетенций, которыми владеет i-й специалист
* Σ - сумма по всем специалистам команды

Помогает понимать насколько специалисту сложно будет работать со всеми компетенциями

* ##### Alice: 1 компетенция, 2 компетенация, 3 компетенция. Доля от всех компетенций, 35% (si) - 0.35, si² - 0.1225
* ##### Bob: 1 компетенция, 2 компетенация. Доля от всех компетенций, 35% (si) - 0.25, si² - 0.0625
* ##### Karl: 1 компетенция, 2 компетенация. Доля от всех компетенций, 5% (si) - 0.05, si² - 0.0025

#### HHI = (0.1225 + 0.0625 + ... + 0.0025) × 10000 = 2500
* HHI < 1500: Низкая концентрация (здоровое распределение)
* 1500 ≤ HHI < 2500: Умеренная концентрация (требует внимания)
* HHI ≥ 2500: Высокая концентрация (критический риск)

### 2.2 Риск информационной асимметрии
Неравномерное распределение критической информации между членами команды, создающее уязвимости в процессах принятия решений и потенциальные конфликты интересов

### Хеджирование
* Внедрение централизованных систем управления знаниями
* Регулярные all-hands встречи с обсуждением ключевых метрик
* Автоматизированные дашборды с real-time метриками производительности

### 2.3  Риск деградации коммуникационных протоколов
Нарушение эффективности информационного обмена между подразделениями, приводящее к задержкам в имплементации критических изменений в торговых алгоритмах

### Хеджирование
* Стандартизация коммуникационных протоколов (RACI-матрица)
* Регулярные анализы коммуникационных сбоев
* Использование специализированных платформ для управления задачами

### 2.4 Риск выгорания и психологической усталости
Хроническое перенапряжение специалистов в условиях высокой волатильности рынков и постоянного давления на производительность, приводящее к снижению качества принимаемых решений и увеличению вероятности критических ошибок

### Хеджирование
* Мониторинг workload и автоматические алерты о перегрузке
* Обязательные периоды отдыха после интенсивных проектов
* Доступ к психологической поддержке и коучингу

Как исследовать? Через опросы

### Maslach Burnout Inventory (MBI)

* Эмоциональное истощение (Emotional Exhaustion - EE) - Чувство эмоционального перенапряжения и опустошения
* Деперсонализация (Depersonalization - DP) - Эмоциональное дистанцирование
* Редукция личных достижений (Personal Accomplishment - PA) - Снижение чувства компетентности

#### Задаются вопросы в опроснике (шкала ответов 0-6):
* EE_score = Σ(EE_questions) / 9 вопросов
* DP_score = Σ(DP_questions) / 5 вопросов
* PA_score = Σ(PA_questions) / 8 вопросов

#### Burnout_Index = (EE_score + DP_score + (6 (макс значение) - PA_score)) / 3

Для себя выбрать пороговый коэффициент и измерять его или
* EE: низкий <2, высокий >3.3
* DPL: низкий <1, высокий >2.3
* PA: низкий >4, выоский <2.9

### 2.5 Риск организационной инерции

## 3.1 Риск корреляции стратегий
В периоды рыночного стресса стратегии, которые исторически показывали низкую корреляцию, могут начать двигаться синхронно. Например, во время резкого падения рынка и momentum, и mean-reversion стратегии могут одновременно генерировать убытки

## Dynamic Conditional Correlation (DCC-GARCH)
Анализ корреляций между торговыми стратегиями. Представьте, что у вас 5 торговых стратегий. Если они все начнут терять деньги одновременно (высокая корреляция), ваш портфель получит большой лосс

#### DCC-GARCH помогает:
* Отслеживать изменения корреляций во времени - корреляции не постоянны, они меняются
* Предсказывать периоды высокого риска - когда стратегии начинают двигаться синхронно
* Оптимизировать веса стратегий - сколько денег выделить каждой стратегии
* Избегать ложной диверсификации - когда кажется, что риски разные, а на деле одинаковые

#### Шаг 1: GARCH для каждой стратегии
* Расчитываем GARCH(1,1) модель отдельно для каждой стратегии
* Извлекаем стандартизированные остатки (остатки/волатильность)

#### Шаг 2: Оценка параметров DCC
* Находим параметры a (реакция на шоки) и b (инерция)
* Использует максимальное правдоподобие с ограничением a+b<1

#### Шаг 3: Динамические корреляции
* Рекурсивно обновляем корреляции: Q(t) = (1-a-b)×Q̄ + a×(остатки×остатки') + b×Q(t-1)
* Нормализуем Q для получения корреляционной матрицы

### Нормальные и опасные значения
1. Нормальные значения:
* Корреляция 0.3-0.7 - стратегии связаны, но не критично
* Волатильность корреляции < 0.2 - связь стабильная
* Параметры DCC: a = 0.01-0.05, b = 0.90-0.98

2. Опасные сигналы:
* Корреляция > 0.8 - стратегии двигаются почти синхронно
Р* езкий рост корреляции (с 0.3 до 0.8 за неделю) - признак кризиса
* a + b > 0.99 - система нестабильна

```python
import numpy as np
import pandas as pd
from arch import arch_model
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

class DCCGarch:
    """
    Класс для расчета Dynamic Conditional Correlation с GARCH моделями
    """
    
    def __init__(self):
        self.garch_models = {}
        self.standardized_residuals = None
        self.dcc_params = None
        self.dynamic_correlations = None
        
    def fit_garch_models(self, returns_df):
        """
        Шаг 1: Подгоняем GARCH(1,1) модель для каждой стратегии
        """
        print("Шаг 1: Подгонка GARCH моделей для каждой стратегии...")
        
        garch_results = {}
        standardized_residuals = pd.DataFrame(index=returns_df.index)
        
        for strategy in returns_df.columns:
            print(f"Обрабатываем стратегию: {strategy}")
            
            # Создаем GARCH(1,1) модель
            model = arch_model(returns_df[strategy] * 100,  # Умножаем на 100 для стабильности
                             vol='Garch', p=1, q=1, 
                             mean='constant', dist='normal')
            
            # Подгоняем модель
            result = model.fit(disp='off')
            garch_results[strategy] = result
            
            # Получаем стандартизированные остатки
            residuals = result.resid / result.conditional_volatility
            standardized_residuals[strategy] = residuals
            
            # Выводим параметры GARCH
            print(f"  α (alpha): {result.params['alpha[1]']:.4f}")
            print(f"  β (beta):  {result.params['beta[1]']:.4f}")
            print(f"  ω (omega): {result.params['omega']:.6f}")
            print()
        
        self.garch_models = garch_results
        self.standardized_residuals = standardized_residuals.dropna()
        
    def estimate_dcc_parameters(self):
        """
        Шаг 2: Оцениваем параметры DCC модели
        """
        print("Шаг 2: Оценка параметров DCC...")
        
        # Вычисляем долгосрочную корреляционную матрицу
        long_term_corr = self.standardized_residuals.corr().values
        
        # Начальные параметры для оптимизации
        initial_params = [0.01, 0.95]  # [a, b]
        
        # Оптимизируем параметры DCC
        def dcc_likelihood(params):
            a, b = params
            if a < 0 or b < 0 or a + b >= 1:
                return 1e6
            
            T = len(self.standardized_residuals)
            n = len(self.standardized_residuals.columns)
            
            Q = long_term_corr.copy()
            log_likelihood = 0
            
            residuals_array = self.standardized_residuals.values
            
            for t in range(1, T):
                # Обновляем Q согласно DCC формуле
                outer_product = np.outer(residuals_array[t-1], residuals_array[t-1])
                Q = (1 - a - b) * long_term_corr + a * outer_product + b * Q
                
                # Нормализуем для получения корреляционной матрицы
                Q_sqrt_inv = np.diag(1.0 / np.sqrt(np.diag(Q)))
                R = Q_sqrt_inv @ Q @ Q_sqrt_inv
                
                # Добавляем к логарифмической правдоподобности
                try:
                    log_likelihood -= 0.5 * (np.log(np.linalg.det(R)) + 
                                           residuals_array[t] @ np.linalg.inv(R) @ residuals_array[t])
                except:
                    return 1e6
                    
            return -log_likelihood
        
        # Оптимизируем
        bounds = [(0.001, 0.3), (0.001, 0.998)]
        constraints = {'type': 'ineq', 'fun': lambda x: 0.999 - sum(x)}
        
        result = minimize(dcc_likelihood, initial_params, 
                         bounds=bounds, constraints=constraints)
        
        self.dcc_params = result.x
        print(f"Оптимальные параметры DCC:")
        print(f"  a (скорость адаптации): {self.dcc_params[0]:.4f}")
        print(f"  b (инерция): {self.dcc_params[1]:.4f}")
        print(f"  a + b = {sum(self.dcc_params):.4f}")
        print()
        
    def calculate_dynamic_correlations(self):
        """
        Шаг 3: Рассчитываем динамические корреляции
        """
        print("Шаг 3: Расчет динамических корреляций...")
        
        a, b = self.dcc_params
        residuals_array = self.standardized_residuals.values
        long_term_corr = self.standardized_residuals.corr().values
        
        T = len(residuals_array)
        n = len(self.standardized_residuals.columns)
        
        # Инициализируем
        Q = long_term_corr.copy()
        correlations = []
        
        for t in range(1, T):
            # DCC уравнение
            outer_product = np.outer(residuals_array[t-1], residuals_array[t-1])
            Q = (1 - a - b) * long_term_corr + a * outer_product + b * Q
            
            # Нормализация
            Q_sqrt_inv = np.diag(1.0 / np.sqrt(np.diag(Q)))
            R = Q_sqrt_inv @ Q @ Q_sqrt_inv
            
            correlations.append(R.copy())
        
        # Сохраняем результаты
        self.dynamic_correlations = correlations
        
        # Создаем DataFrame с корреляциями для удобства
        strategy_names = self.standardized_residuals.columns
        dates = self.standardized_residuals.index[1:]  # Убираем первую дату
        
        # Извлекаем корреляции между всеми парами стратегий
        correlation_series = {}
        for i in range(n):
            for j in range(i+1, n):
                pair_name = f"{strategy_names[i]}_vs_{strategy_names[j]}"
                pair_correlations = [corr[i,j] for corr in correlations]
                correlation_series[pair_name] = pd.Series(pair_correlations, index=dates)
        
        self.correlation_df = pd.DataFrame(correlation_series)
        print("Динамические корреляции рассчитаны!")
        
    def analyze_correlations(self):
        """
        Анализ результатов и выявление рисков
        """
        print("\n" + "="*50)
        print("АНАЛИЗ РИСКОВ КОРРЕЛЯЦИИ")
        print("="*50)
        
        for pair in self.correlation_df.columns:
            corr_series = self.correlation_df[pair]
            
            print(f"\nПара стратегий: {pair}")
            print(f"  Средняя корреляция: {corr_series.mean():.3f}")
            print(f"  Максимальная корреляция: {corr_series.max():.3f}")
            print(f"  Минимальная корреляция: {corr_series.min():.3f}")
            print(f"  Волатильность корреляции: {corr_series.std():.3f}")
            
            # Анализ рисков
            if corr_series.max() > 0.8:
                print(f"  ⚠️  ВЫСОКИЙ РИСК: Максимальная корреляция > 0.8")
            
            if corr_series.std() > 0.2:
                print(f"  ⚠️  НЕСТАБИЛЬНОСТЬ: Высокая волатильность корреляции")
                
            # Считаем, сколько времени корреляция была опасно высокой
            high_corr_pct = (corr_series > 0.7).mean() * 100
            if high_corr_pct > 20:
                print(f"  ⚠️  ЧАСТЫЕ КРИЗИСЫ: {high_corr_pct:.1f}% времени корреляция > 0.7")
    
    def plot_results(self):
        """
        Построение графиков результатов
        """
        fig, axes = plt.subplots(2, 1, figsize=(12, 10))
        
        # График 1: Динамические корреляции
        self.correlation_df.plot(ax=axes[0], alpha=0.7)
        axes[0].set_title('Динамические корреляции между стратегиями')
        axes[0].set_ylabel('Корреляция')
        axes[0].axhline(y=0.7, color='red', linestyle='--', alpha=0.5, label='Опасный уровень (0.7)')
        axes[0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        axes[0].grid(True, alpha=0.3)
        
        # График 2: Средние корреляции по периодам
        monthly_corr = self.correlation_df.resample('M').mean()
        monthly_corr.plot(kind='bar', ax=axes[1], alpha=0.7)
        axes[1].set_title('Средние месячные корреляции')
        axes[1].set_ylabel('Корреляция')
        axes[1].tick_params(axis='x', rotation=45)
        
        plt.tight_layout()
        plt.show()

# Пример использования
def create_sample_data():
    """
    Создаем примерные данные для 5 стратегий
    """
    np.random.seed(42)
    dates = pd.date_range('2020-01-01', '2023-12-31', freq='D')
    
    # Базовые доходности с разными характеристиками
    n_days = len(dates)
    
    # Стратегия 1: Momentum (тренд-следование)
    momentum = np.random.normal(0.0005, 0.015, n_days)
    
    # Стратегия 2: Mean Reversion (возврат к среднему)
    mean_reversion = np.random.normal(0.0003, 0.012, n_days)
    
    # Стратегия 3: Carry Trade (валютный кэрри)
    carry = np.random.normal(0.0004, 0.018, n_days)
    
    # Стратегия 4: Statistical Arbitrage (стат. арбитраж)
    stat_arb = np.random.normal(0.0002, 0.008, n_days)
    
    # Стратегия 5: Volatility Trading (торговля волатильностью)
    vol_trading = np.random.normal(0.0001, 0.020, n_days)
    
    # Добавляем общие шоки (кризисы) для создания корреляции
    common_shocks = np.random.normal(0, 0.01, n_days)
    crisis_periods = np.random.binomial(1, 0.05, n_days) * np.random.normal(0, 0.03, n_days)
    
    # Применяем шоки с разной силой к каждой стратегии
    momentum += 0.7 * common_shocks + 0.8 * crisis_periods
    mean_reversion += 0.5 * common_shocks + 0.6 * crisis_periods  
    carry += 0.8 * common_shocks + 0.9 * crisis_periods
    stat_arb += 0.3 * common_shocks + 0.4 * crisis_periods
    vol_trading += -0.4 * common_shocks + 0.5 * crisis_periods  # Обратная корреляция в обычное время
    
    # Создаем DataFrame
    returns_df = pd.DataFrame({
        'Momentum': momentum,
        'Mean_Reversion': mean_reversion,
        'Carry_Trade': carry,
        'Stat_Arbitrage': stat_arb,
        'Vol_Trading': vol_trading
    }, index=dates)
    
    return returns_df

# Основной код для демонстрации
if __name__ == "__main__":
    print("Демонстрация DCC-GARCH для анализа корреляционного риска стратегий")
    print("="*70)
    
    # Создаем данные
    print("Создаем примерные данные для 5 торговых стратегий...")
    returns_data = create_sample_data()
    
    print(f"Данные созданы: {len(returns_data)} наблюдений с {returns_data.index[0].strftime('%Y-%m-%d')} по {returns_data.index[-1].strftime('%Y-%m-%d')}")
    print("\nОписательная статистика доходностей:")
    print(returns_data.describe())
    
    # Инициализируем DCC-GARCH модель
    dcc_model = DCCGarch()
    
    # Выполняем анализ
    dcc_model.fit_garch_models(returns_data)
    dcc_model.estimate_dcc_parameters()  
    dcc_model.calculate_dynamic_correlations()
    dcc_model.analyze_correlations()
    
    # Строим графики
    dcc_model.plot_results()
    
    print("\n" + "="*50)
    print("РЕКОМЕНДАЦИИ ПО УПРАВЛЕНИЮ РИСКАМИ")
    print("="*50)
    
    # Анализируем текущее состояние
    latest_correlations = dcc_model.correlation_df.iloc[-1]
    max_current_corr = latest_correlations.max()
    
    if max_current_corr > 0.8:
        print("🚨 КРИТИЧЕСКИЙ УРОВЕНЬ РИСКА!")
        print("   Рекомендации:")
        print("   • Немедленно снизить размер позиций на 50%")
        print("   • Добавить защитные позиции (VIX, золото)")
        print("   • Увеличить кэш до 20-30% портфеля")
        
    elif max_current_corr > 0.6:
        print("⚠️  ПОВЫШЕННЫЙ УРОВЕНЬ РИСКА")
        print("   Рекомендации:")
        print("   • Снизить размер позиций на 20%")
        print("   • Добавить некоррелированные активы")
        print("   • Мониторить ситуацию ежедневно")
        
    else:
        print("✅ НОРМАЛЬНЫЙ УРОВЕНЬ РИСКА")
        print("   • Текущая диверсификация работает")
        print("   • Продолжить регулярный мониторинг")
    
    print(f"\nТекущая максимальная корреляция: {max_current_corr:.3f}")
    print("Модель готова для использования в торговой системе!")
```

## 3.2 Риск деградации альфы портфеля
Риск деградации альфы портфеля — это вероятность того, что способность инвестиционной стратегии генерировать избыточную доходность (альфу) со временем ухудшится или исчезнет. Альфа представляет собой доходность портфеля сверх того, что можно было бы ожидать исходя из его рыночного риска (беты). Это показатель мастерства управляющего или эффективности инвестиционной стратегии

## CUSUM (Cumulative Sum) — это статистический метод, который помогает вовремя заметить, когда ваша торговая стратегия перестает работать так же хорошо, как раньше

Представьте ситуацию: у вас есть торговая стратегия, которая в среднем приносила 2% в месяц. Но рынок меняется, и стратегия начинает приносить только 0.5% или даже уходит в минус. Проблема в том, что из-за естественной волатильности доходности это изменение может быть незаметно месяцами! CUSUM помогает отличить временную просадку от фундаментального ухудшения стратегии

Основные шаги алгоритма CUSUM:

### 1. Установка параметров
* Определяется ожидаемая доходность (mu) и волатильность (sigma) стратегии
* Устанавливается порог чувствительности k = 0.5 × sigma
* Устанавливается критический порог h = 4 × sigma

### 2. Накопление отклонений
* Для каждого периода вычисляется отклонение: текущая доходность минус ожидаемая доходность минус порог k
* Если отклонение отрицательное (стратегия работает хуже ожидаемого), оно добавляется к накопленной сумме
* Если накопленная сумма становится отрицательной, она обнуляется (CUSUM не может быть < 0)

### 3. Проверка на деградацию
* Когда накопленная сумма превышает критический порог h, срабатывает сигнал тревоги
* Это означает, что стратегия стабильно показывает результаты хуже ожидаемых

### 4. Реакция на сигналы
* При обнаружении деградации система уменьшает выделенный капитал на стратегию вдвое
* Это позволяет минимизировать потери от деградирующих стратегий

#### Ключевая идея:
CUSUM отслеживает накопленные негативные отклонения от ожидаемой производительности. Когда эти отклонения становятся слишком большими, алгоритм сигнализирует о вероятной деградации стратегии.

### Нормальные и критические значения
#### Нормальные значения:
* CUSUM колеблется около 0 или медленно растет до 1-2σ
* Периодические сбросы к нулю — это нормально

#### Тревожные сигналы:
* CUSUM превышает 4σ — вероятная деградация стратегии
* Быстрый рост CUSUM (за 5-10 периодов) — резкое ухудшение
* CUSUM не сбрасывается к нулю месяцами — устойчивое снижение альфы

#### Что делать при превышении порога:
* Уменьшить размер позиции на 50%
* Провести анализ причин деградации
* Если через 5-10 периодов ситуация не улучшается — остановить стратегию
* Перенастроить параметры или заменить стратегию

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Создаем симуляцию доходностей для 5 стратегий
np.random.seed(42)
n_periods = 100  # количество торговых периодов

# Параметры стратегий (ожидаемая доходность и волатильность)
strategies = {
    'Momentum': {'mu': 0.02, 'sigma': 0.05, 'degradation_start': 60},
    'Mean Reversion': {'mu': 0.015, 'sigma': 0.03, 'degradation_start': None},
    'Pairs Trading': {'mu': 0.01, 'sigma': 0.02, 'degradation_start': 40},
    'Trend Following': {'mu': 0.025, 'sigma': 0.06, 'degradation_start': 80},
    'Arbitrage': {'mu': 0.005, 'sigma': 0.01, 'degradation_start': 30}
}

def generate_returns(mu, sigma, n_periods, degradation_start=None):
    """Генерирует доходности с возможной деградацией"""
    returns = np.random.normal(mu, sigma, n_periods)
    
    # Добавляем деградацию если указано
    if degradation_start is not None:
        degradation_factor = np.linspace(1, 0.2, n_periods - degradation_start)
        returns[degradation_start:] = returns[degradation_start:] * degradation_factor
        # Добавляем дополнительный шум при деградации
        returns[degradation_start:] += np.random.normal(0, sigma * 0.5, n_periods - degradation_start)
    
    return returns

def calculate_cusum(returns, mu, sigma, k_factor=0.5, h_factor=4):
    """
    Рассчитывает CUSUM для обнаружения деградации
    
    Parameters:
    - returns: массив доходностей
    - mu: ожидаемая средняя доходность
    - sigma: стандартное отклонение
    - k_factor: коэффициент для порога чувствительности (обычно 0.5)
    - h_factor: коэффициент для критического порога (обычно 4-5)
    """
    k = k_factor * sigma  # порог чувствительности
    h = h_factor * sigma  # критический порог
    
    cusum = np.zeros(len(returns))
    alerts = []
    
    for t in range(1, len(returns)):
        # Отклонение от ожидаемого с учетом порога
        z_t = returns[t] - mu - k
        
        # CUSUM не может быть отрицательным
        cusum[t] = max(0, cusum[t-1] + z_t)
        
        # Проверяем превышение критического порога
        if cusum[t] > h:
            alerts.append(t)
    
    return cusum, alerts, h

# Генерируем данные и рассчитываем CUSUM для каждой стратегии
results = {}
for name, params in strategies.items():
    returns = generate_returns(
        params['mu'], 
        params['sigma'], 
        n_periods, 
        params['degradation_start']
    )
    
    cusum, alerts, threshold = calculate_cusum(
        returns, 
        params['mu'], 
        params['sigma']
    )
    
    results[name] = {
        'returns': returns,
        'cusum': cusum,
        'alerts': alerts,
        'threshold': threshold,
        'cumulative_returns': np.cumprod(1 + returns) - 1
    }

# Визуализация результатов
fig, axes = plt.subplots(5, 2, figsize=(15, 20))
fig.suptitle('Анализ деградации альфы методом CUSUM для 5 стратегий', fontsize=16)

for idx, (name, data) in enumerate(results.items()):
    # График доходности
    ax1 = axes[idx, 0]
    ax1.plot(data['cumulative_returns'] * 100, label='Накопленная доходность')
    ax1.set_title(f'{name} - Доходность')
    ax1.set_xlabel('Период')
    ax1.set_ylabel('Доходность (%)')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # График CUSUM
    ax2 = axes[idx, 1]
    ax2.plot(data['cusum'], label='CUSUM', color='blue')
    ax2.axhline(y=data['threshold'], color='red', linestyle='--', label=f'Порог = {data["threshold"]:.3f}')
    
    # Отмечаем точки тревоги
    if data['alerts']:
        ax2.scatter(data['alerts'], [data['cusum'][i] for i in data['alerts']], 
                   color='red', s=50, label='Сигнал деградации')
    
    ax2.set_title(f'{name} - CUSUM анализ')
    ax2.set_xlabel('Период')
    ax2.set_ylabel('CUSUM')
    ax2.grid(True, alpha=0.3)
    ax2.legend()

plt.tight_layout()
plt.show()

# Выводим статистику
print("=== СТАТИСТИКА ДЕГРАДАЦИИ ===\n")
for name, data in results.items():
    degradation_period = strategies[name]['degradation_start']
    first_alert = data['alerts'][0] if data['alerts'] else None
    
    print(f"{name}:")
    print(f"  Ожидаемая доходность: {strategies[name]['mu']*100:.1f}%")
    print(f"  Волатильность: {strategies[name]['sigma']*100:.1f}%")
    print(f"  Начало деградации: период {degradation_period if degradation_period else 'Нет'}")
    print(f"  Первый сигнал CUSUM: период {first_alert if first_alert else 'Нет'}")
    
    if first_alert and degradation_period:
        delay = first_alert - degradation_period
        print(f"  Задержка обнаружения: {delay} периодов")
    
    print(f"  Финальная доходность: {(data['cumulative_returns'][-1]*100):.1f}%")
    print()

# Пример адаптивного управления капиталом
print("=== АДАПТИВНОЕ УПРАВЛЕНИЕ КАПИТАЛОМ ===\n")

initial_capital = 1000000  # начальный капитал
allocations = {name: initial_capital / 5 for name in strategies.keys()}  # равные доли

for period in range(n_periods):
    # Проверяем CUSUM для каждой стратегии
    for name, data in results.items():
        if period in data['alerts']:
            # Уменьшаем аллокацию на 50% при сигнале
            allocations[name] *= 0.5
            print(f"Период {period}: Уменьшена аллокация на {name} до ${allocations[name]:,.0f}")

# Финальные аллокации
print(f"\nФинальные аллокации:")
for name, allocation in allocations.items():
    print(f"  {name}: ${allocation:,.0f} ({allocation/sum(allocations.values())*100:.1f}%)")
```

### 3.3 Риск аллокации капитала в портфеле

Риск концентрации возникает, когда слишком большая доля капитала выделена одной стратегии или группе коррелированных стратегий. Это может привести к значительным потерям при неблагоприятном развитии событий

* #### 0 - Создавайте свой метод для оценки аллокации капитала
* #### 1 - Сделать равную аллокацию
* #### 2 - Уменьшить аллокацию по дродауну из бектеста
* #### 3 - Модель Марковица (Mean-Variance Optimization)
Находит оптимальное распределение активов в портфеле, балансируя между ожидаемой доходностью и риском. Метод строит "эффективную границу" - множество портфелей с максимальной доходностью для каждого уровня риска

Как работает: Использует исторические данные для расчета ожидаемой доходности активов, их волатильности (стандартного отклонения) и корреляций между ними. Затем решает задачу квадратичной оптимизации: максимизировать доходность при заданном уровне риска или минимизировать риск при заданной доходности.

* #### 4 - CVaR оптимизация (Conditional Value at Risk)
Минимизирует средние потери в худших сценариях (например, в 5% самых неблагоприятных случаев). В отличие от VaR, который показывает пороговое значение потерь, CVaR измеряет среднюю величину потерь за этим порогом

Как работает: Вычисляет среднее значение потерь, превышающих VaR на заданном уровне доверия. Например, если VaR на уровне 95% составляет -10%, CVaR покажет средние потери в тех 5% случаев, когда потери превысили 10%

* #### 5 - Kelly Criterion
Определяет оптимальную долю капитала для инвестирования в актив с положительным математическим ожиданием, максимизируя долгосрочный темп роста капитала

Формула: f* = (p × b - q) / b, где f* - оптимальная доля капитала, p - вероятность выигрыша (win), q - вероятность проигрыша (loss), b - коэффициент выигрыша 
Как работает: Балансирует между агрессивным ростом и защитой от разорения. При слишком малых ставках рост медленный, при слишком больших - высок риск больших потерь

* #### 6 - Risk Parity
Распределяет капитал между активами так, чтобы каждая стратегия вносила одинаковый вклад в общий риск портфеля, а не просто имела равную долю по стоимости
Как работает: Вместо равных весов по капиталу, метод выравнивает вклад в риск. Поскольку BTC менее волатилен, его доля в портфеле будет больше, чтобы вклад в общий риск равнялся аллокации например в SOL

# 4. Риск на уровне стратегии

### 4.1 Риск переобучения (Overfitting Risk)
Чрезмерная оптимизация параметров модели на исторических данных, приводящая к деградации производительности на out-of-sample данных. Модель захватывает шум вместо истинных рыночных паттернов

### Sharp Ratio
Показывает, какую избыточную доходность приносит стратегия на единицу принятого риска.
Sharpe Ratio = (Rp - Rf) / σp, где:

* Rp — средняя доходность стратегии за период
* Rf — безрисковая ставка доходности
* σp — стандартное отклонение доходности стратегии

## Deflated Sharpe Ratio (DSR)

```python
from IPython.display import Image

Image(url='https://gmarti.gitlab.io/assets/deflated_sharpe_ratio.png')
```

Это стат метрика, разработанная для решения проблемы переоценки эффективности торговых стратегий из-за множественного тестирования и data mining
Оценить вероятность того, что наблюдаемый коэффициент Шарпа стратегии является статистически значимым, а не результатом случайности или переподгонки (overfitting)

Представьте, что вы тестируете 1000 разных настроек стратегии. Хотя бы одна из них покажет отличный результат просто случайно - как если бы 1000 человек подбрасывали монетку, и кто-то выбросил бы 10 орлов подряд. Это не значит, что у него есть суперспособность - просто повезло

### Зачем нужен DSR:
* При разработке алгостратегий мы часто тестируем множество параметров
* Чем больше вариантов мы проверяем, тем выше шанс найти "случайно успешную" стратегию
* DSR показывает, насколько результат стратегии статистически значим с учетом количества попыток

### Как работает DSR:
* Считаем обычный Sharpe Ratio стратегии
* Вычисляем SR* - максимальный Sharpe Ratio, который можно получить случайно при тестировании N стратегий
* Сравниваем наш SR с SR* через Z-score или масштабируем на стандартное отклонение
* Получаем DSR - вероятность того, что результат не случаен

Итог:

* DSR < 0.5 = скорее всего переподгонка
* DSR > 0.95 = скорее всего реальный результат

## Probability of Backtest Overfitting (PBO)
Это инструмент для оценки вероятности того, что ваша торговая стратегия показывает хорошие результаты просто случайно, а не потому что она действительно работает. PBO показывает вероятность того, что ваша стратегия "переобучилась" на исторических данных

#### Для чего нужен:
* Фильтрация плохих стратегий до запуска с реальными деньгами
* Сравнение разных стратегий между собой
* Оценка устойчивости стратегии к изменениям рынка

### 1. Разделение данных
* Делим исторические данные пополам: IS (первые 50%) и OOS (вторые 50%)

### 2. Расчет Sharpe Ratio
* Для каждой стратегии считаем Sharpe отдельно для IS и OOS периодов
* Sharpe = (Средняя доходность / Стандартное отклонение) × √365

### 3. Ранжирование
* Ранжируем стратегии по Sharpe (лучшая = ранг 0)
* Отдельно для IS и OOS периодов

### 4. Анализ пар
* Создаем все возможные пары стратегий (для 10 стратегий = 45 пар)
* Для каждой пары сравниваем относительные ранги в IS и OOS

### 5. Проверка согласованности
* Если знаки относительных рангов одинаковые → пара согласована ✅
* Если знаки разные → пара переоптимизирована ❌

### 6. Расчет PBO
* PBO = Количество переоптимизированных пар / Общее количество пар

### Интерпретация:
* PBO < 30% → Низкий риск переоптимизации ✅
* PBO 30-50% → Умеренный риск ⚠️
* PBO > 50% → Высокий риск переоптимизации ❌

```python
import numpy as np
import pandas as pd
from itertools import combinations
import matplotlib.pyplot as plt

def calculate_pbo(returns_matrix, split_ratio=0.5):
    """
    Расчет вероятности переоптимизации (PBO)
    
    Параметры:
    returns_matrix: матрица доходностей (строки - время, столбцы - варианты стратегии)
    split_ratio: доля данных для in-sample (обычно 0.5)
    
    Возвращает:
    pbo: вероятность переоптимизации
    """
    
    # Шаг 1: Разбиваем данные на IS и OOS
    n_periods = returns_matrix.shape[0]
    split_point = int(n_periods * split_ratio)
    
    is_returns = returns_matrix[:split_point]  # In-Sample данные
    oos_returns = returns_matrix[split_point:]  # Out-of-Sample данные
    
    # Шаг 2: Считаем метрику производительности (Sharpe Ratio)
    def sharpe_ratio(returns):
        """Расчет коэффициента Шарпа"""
        if len(returns) == 0:
            return 0
        mean_return = np.mean(returns)
        std_return = np.std(returns)
        if std_return == 0:
            return 0
        return mean_return / std_return * np.sqrt(365)  # Годовой Sharpe
    
    # Считаем Sharpe для каждого варианта стратегии
    n_strategies = returns_matrix.shape[1]
    is_sharpe = np.array([sharpe_ratio(is_returns[:, i]) for i in range(n_strategies)])
    oos_sharpe = np.array([sharpe_ratio(oos_returns[:, i]) for i in range(n_strategies)])
    
    # Шаг 3: Считаем ранги (чем выше Sharpe, тем лучше ранг)
    is_ranks = np.argsort(np.argsort(-is_sharpe))  # Ранги по убыванию
    oos_ranks = np.argsort(np.argsort(-oos_sharpe))
    
    # Шаг 4: Считаем PBO
    # Создаем все возможные пары стратегий
    n_pairs = 0
    n_overfitted = 0
    
    for i in range(n_strategies):
        for j in range(i + 1, n_strategies):
            n_pairs += 1
            
            # Проверяем, согласуются ли ранги
            is_relative = is_ranks[i] - is_ranks[j]
            oos_relative = oos_ranks[i] - oos_ranks[j]
            
            # Если знаки разные, это признак переоптимизации
            if is_relative * oos_relative < 0:
                n_overfitted += 1
    
    # Вычисляем PBO
    pbo = n_overfitted / n_pairs if n_pairs > 0 else 0
    
    return pbo, is_sharpe, oos_sharpe

# ПРИМЕР ИСПОЛЬЗОВАНИЯ
np.random.seed(42)

# Создаем синтетические данные для 10 вариантов стратегии
n_days = 500  # Количество торговых дней
n_strategies = 10  # Количество вариантов стратегии

# Генерируем доходности
# Некоторые стратегии будут переоптимизированными
returns_matrix = np.zeros((n_days, n_strategies))

for i in range(n_strategies):
    if i < 3:  # Первые 3 стратегии - хорошие (работают стабильно)
        # Стабильная положительная доходность
        returns_matrix[:, i] = np.random.normal(0.0005, 0.01, n_days)
    else:  # Остальные - переоптимизированные
        # Хорошо работают в первой половине, плохо во второй
        returns_matrix[:n_days//2, i] = np.random.normal(0.001, 0.01, n_days//2)
        returns_matrix[n_days//2:, i] = np.random.normal(-0.0005, 0.015, n_days//2)

# Рассчитываем PBO
pbo, is_sharpe, oos_sharpe = calculate_pbo(returns_matrix)

print(f"Вероятность переоптимизации (PBO): {pbo:.2%}")
print("\nИнтерпретация:")
if pbo < 0.3:
    print("✅ Отличный результат! Стратегии скорее всего не переоптимизированы")
elif pbo < 0.5:
    print("⚠️ Приемлемый результат, но есть риск переоптимизации")
else:
    print("❌ Высокий риск переоптимизации! Стратегии вряд ли будут работать на реальном рынке")

# Визуализация результатов
plt.figure(figsize=(12, 5))

# График 1: Сравнение Sharpe Ratio
plt.subplot(1, 2, 1)
x = np.arange(n_strategies)
width = 0.35

plt.bar(x - width/2, is_sharpe, width, label='In-Sample', alpha=0.8)
plt.bar(x + width/2, oos_sharpe, width, label='Out-of-Sample', alpha=0.8)
plt.xlabel('Номер стратегии')
plt.ylabel('Sharpe Ratio')
plt.title('Сравнение производительности стратегий')
plt.legend()
plt.grid(True, alpha=0.3)

# График 2: Кумулятивная доходность
plt.subplot(1, 2, 2)
cumulative_returns = (1 + returns_matrix).cumprod(axis=0)
for i in range(min(5, n_strategies)):  # Показываем первые 5 стратегий
    plt.plot(cumulative_returns[:, i], label=f'Стратегия {i+1}', alpha=0.7)

plt.axvline(x=n_days//2, color='red', linestyle='--', label='IS/OOS разделение')
plt.xlabel('Торговые дни')
plt.ylabel('Кумулятивная доходность')
plt.title('Динамика стратегий во времени')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Дополнительная статистика
print("\nДетальная статистика по стратегиям:")
print("-" * 60)
print(f"{'Стратегия':>10} | {'IS Sharpe':>10} | {'OOS Sharpe':>10} | {'Статус':>15}")
print("-" * 60)

for i in range(n_strategies):
    status = "Хорошая" if is_sharpe[i] > 0 and oos_sharpe[i] > 0 and abs(is_sharpe[i] - oos_sharpe[i]) < 1 else "Переоптимизир."
    print(f"{i+1:>10} | {is_sharpe[i]:>10.2f} | {oos_sharpe[i]:>10.2f} | {status:>15}")

# Функция для оценки новой стратегии
def evaluate_strategy(returns, strategy_name="Моя стратегия"):
    """Оценка одной стратегии на переоптимизацию"""
    
    # Создаем случайные вариации стратегии для сравнения
    n_variations = 20
    variations = np.zeros((len(returns), n_variations))
    
    # Оригинальная стратегия
    variations[:, 0] = returns
    
    # Создаем вариации с небольшими изменениями
    for i in range(1, n_variations):
        noise = np.random.normal(0, 0.001, len(returns))
        variations[:, i] = returns + noise
    
    pbo, _, _ = calculate_pbo(variations)
    
    print(f"\nОценка стратегии '{strategy_name}':")
    print(f"PBO = {pbo:.2%}")
    
    if pbo < 0.5:
        print("✅ Стратегия прошла тест на переоптимизацию")
    else:
        print("❌ Стратегия не прошла тест, высокий риск переоптимизации")
    
    return pbo

# Пример оценки новой стратегии
test_returns = np.random.normal(0.0003, 0.01, 250)  # Тестовая стратегия
evaluate_strategy(test_returns, "Тестовая стратегия")
```

### Хеджирование
Кросс-валидация с временными рядами

### Walk-forward analysis

Метод последовательного тестирования, который имитирует реальную торговлю. Работает следующим образом:

* Разделение данных: Весь датасет делится на окна обучения и тестирования, которые последовательно сдвигаются во времени
* Обучение: Модель обучается на первом окне данных (например, 1 год)
* Тестирование: Модель тестируется на следующем периоде (например, 6 месяцев)
* Сдвиг окна: Окно сдвигается, и процесс повторяется

#### Преимущества: Учитывает временную структуру данных, предотвращает заглядывание в будущее, близок к реальной торговле.
#### Недостатки: Использует данные неэффективно (каждая точка используется только один раз для теста), может давать нестабильные результаты при малом количестве данных

### K-fold

Классический метод машинного обучения, адаптированный для временных рядов:

* Разделение: Данные делятся на K примерно равных частей (фолдов)
* Итерации: Проводится K итераций, где каждый фолд по очереди становится тестовым
* Обучение: На каждой итерации модель обучается на K-1 фолдах
* Усреднение: Результаты усредняются по всем итерациям

# VaR (Value at Risk)
* VaR - это статистическая мера, которая оценивает максимальный потенциальный убыток портфеля за определенный период времени с заданным уровнем доверия. Например, дневной/месячный VaR на уровне 95% показывает максимальный убыток, который не будет превышен в 95% случаев

#### Для нормального распределения:

* μ - среднее значение доходности
* σ - стандартное отклонение доходности
* z_α - квантиль стандартного нормального распределения
* α - уровень доверия (например, 0.95)

# 4.4 Риск расхождения лайва и бектеста

### Почему результаты расходятся?
#### 1. Проскальзывание (slippage):
* В бектесте: покупаете точно по цене из истории

#### 2. Комиссии
* Часто в бектесте забывают учесть все виды комиссий
* В реальности: комиссия брокера + биржевая комиссия + спред

#### 3. Ликвидность
* В бектесте: можете "купить" любой объем по исторической цене
* В реальности: большой ордер двигает цену против вас

#### 4. Переподгонка (overfitting)
* Стратегия слишком хорошо "выучила" исторические данные
* На новых данных работает хуже

#### 5. Задержки исполнения
* В бектесте: мгновенное исполнение
* В реальности: задержки сети, обработки ордеров

## Модель деградации Шарпа Пуара (Poire)
Эта модель предсказывает, насколько ухудшится коэффициент Шарпа при переходе от бектеста к лайву

SR_live = SR_backtest × √(1 - k), где:
* SR_live - ожидаемый Шарп в реальной торговле
* SR_backtest - Шарп на бектесте
* k - коэффициент деградации (обычно 0.25-0.5)

## Байесовский подход
Использует теорему Байеса для оценки вероятности того, что стратегия действительно прибыльная. Основная идея: обновляем наше мнение о стратегии по мере получения новых данных

#### SR_истинный (Sharp Ratio) = (SR_backtest × n + SR_prior × τ) / (n + τ), где:
* n - количество наблюдений в бектесте
* SR_prior - наше изначальное ожидание Шарпа (обычно 0)
* τ - сила нашего изначального убеждения (обычно 50-100)

#### Пример:
* SR_backtest = 2.5
* n = 250 дней
* SR_prior = 0
* τ = 100

SR_истинный = (2.5 × 250 + 0 × 100) / (250 + 100) = 1.79

- Scoring Model

Score = security + liquidity + regulatory + operational + reputation

#### 1 - Security Score (30% веса)

security = 0.3×PoR + 0.2×Insurance + 0.2×ColdStorage + 0.2×Audit + 0.1×BugBounty
* Proof of Reserves: Соотношение резервов к обязательствам (target >100%)

* Глубина стакана на ±2% от цены
* Соотношение объема к TVL

#### 3 - Regulatory Score (вес 20%)

regulatory = 0.5×Licenses + 0.3×Compliance + 0.2×Transparency
* Наличие лицензий (особенно в ключевых юрисдикциях)
* KYC/AML процедуры
* Публичность команды и юридической структуры

#### 4 - Operational Score (вес 15%)

operational = 0.3×Uptime + 0.5×WithdrawalSpeed + 0.2×Support
* Uptime = (Общее_время - Время_недоступности) / Общее_время × 100% (UptimeRobot, Pingdom, пинг api самостоятельный)

## 1 - Price Impact, насколько ваш ордер "продавит" рынок
* Price Impact (%) = (Execution Price - Mid Price) / Mid Price × 100
* Mid Price = (Best Bid + Best Ask) / 2
* Execution Price = средневзвешенная цена исполнения вашего ордера

#### Пример:
Asks (продажа):
* 30,001 - 2 BTC
* 30,002 - 3 BTC  
* 30,003 - 3 BTC
* $30,005 - 4 BTC

Bids (покупка):
* 30,000 - 2 BTC
* 29,999 - 3 BTC

#### Расчет:

Mid Price = (30,000 + 30,001) / 2 = $30,000.5

Для покупки 10 BTC:
* 2 BTC по 30,001 = 60,002
* 3 BTC по 30,002 = 90,006
* 3 BTC по 30,003 = 90,009
* 2 BTC по 30,005 = 60,010
  
Итого: 300,027 за 10 BTC

#### Average Execution Price = $300,027 / 10 = $30,002.7
#### Price Impact = (30,002.7 - 30,000.5) / 30,000.5 × 100 = 0.0073%


## 2 - Метрика глубины: 2% Market Depth, сколько можно купить/продать при движении цены на 2%

## 3 - Проскальзование (Slippage)

* Фактическая цена = средневзвешенная цена исполнения всего ордера

### 5.3 Риск ликвидации (Liquidation Risk)
Резкие непредсказуемые изменения цены актива, превышающие исторические параметры и заложенные в модель предположения о распределении доходностей.
### Хеджирование
* Динамическое управление размером плеча на основе волатильности и ликвидности
* Многоуровневая система stop-loss с учетом вероятности проскальзывания
* Поддержание буфера маржинального обеспечения выше критических уровней

### Что здесь важно?
* Если у вас есть усреднительная модель, то нужно понять какая у вас средняя цена
* После подставить плечо и высчитать при каком падении у вас будет ликвидация и где

#### Рекомендация: добавлять в баланс больше средств, торговать 80% от этих средств, использовать стоп-лосс, пренудительное закрытие позиции при собственном мониторинге и определение плеча не больше 5x

### 5.4 Риск латентности исполнения (Execution Latency Risk)
Временные задержки между генерацией торгового сигнала и фактическим исполнением ордера, критичные для высокочастотных и арбитражных стратегий.
### Хеджирование
* Колокация серверов в непосредственной близости от биржевых датацентров
* Использование специализированных низколатентных сетевых протоколов
* Параллелизация вычислений и оптимизация кода для минимизации задержек
