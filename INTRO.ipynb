# Что такое стохастичность в контексте алгостратегий
Стохастичность — это случайность в поведении систем. В алготрейдинге:
* Цены активов и рыночные движения не детерминированы (непредсказуемы с точностью), они подвержены случайным флуктуациям
* Результаты сделок в одной и той же стратегии при разных рыночных условиях (или даже с одним и тем же сетапом) могут отличаться
* Маршруты выполнения ордеров, проскальзывание, задержки, влияние других участников рынка — всё это вносит элементы неопределенности

# Выход есть!
Методы оценки стохастичности и устойчивости стратегии
* ### 1. Монте-Карло симуляции
Используется для оценки надежности и устойчивости стратегии к случайным факторам

* ### 2. Bootstrap-анализ
Помогает понять, как стратегия ведет себя на разных возможных "версиях" рынка

* ### 3. Stress-тестирование
Подстановка экстремальных значений (волн, проскальзываний, задержек и т.д.). Показывает, насколько стохастическая изменчивость может "сломать" стратегию

* ### 4. Анализ распределений доходности
Оценка среднего, дисперсии и тд. Проверка на нормальность или наличие "жирных хвостов". Помогает выявить скрытые риски
