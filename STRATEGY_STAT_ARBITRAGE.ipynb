```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from binance.client import Client
from datetime import datetime, timedelta
import time
from itertools import combinations
from scipy import stats
from statsmodels.tsa.stattools import coint, adfuller
from sklearn.linear_model import LinearRegression
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# =============================================================================
# НАСТРОЙКИ АНАЛИЗА
# =============================================================================

# Временные параметры
START_DATE = '2025-01-01'          # Начальная дата анализа
END_DATE = '2025-01-15'            # Конечная дата анализа
TIMEFRAME = '5m'                   # Таймфрейм данных

# Список активов для анализа (фьючерсные символы)
SYMBOLS = [
    'BTCUSDT', 'ETHUSDT', 'SOLUSDT', 'XRPUSDT', 'ARBUSDT', 
    'SEIUSDT', 'SUIUSDT', 'DOGEUSDT', 'ADAUSDT', 'AAVEUSDT',
    'LINKUSDT', 'OPUSDT', 'ENAUSDT', 'AVAXUSDT', 'BCHUSDT',
    'NEARUSDT', 'LTCUSDT', 'GALAUSDT', 'TIAUSDT', 'UNIUSDT',
    'WLDUSDT', 'INJUSDT', 'DOTUSDT', 'APTUSDT'
]

# Критерии отбора пар
TOP_PAIRS_COUNT = 20               # Количество топ пар по корреляции
MIN_CORRELATION = 0.85             # Минимальная корреляция для отбора
COINTEGRATION_P_VALUE = 0.05       # P-value для тестов коинтеграции

# Параметры стационарности
ADF_P_VALUE = 0.05                 # P-value для теста ADF на стационарность

# Параметры скользящей корреляции
ROLLING_WINDOW = 96                # Окно в периодах (96 = 8 часов на 5-минутках)

# Параметры скользящей беты и Z-Score
LOOKBACK_PERIOD = 96               # Период для расчета скользящей беты и Z-Score

# Параметры торговых сигналов
ENTRY_THRESHOLD = 2.0              # Z-Score для входа в позицию
EXIT_THRESHOLD = 0.5               # Z-Score для выхода из позиции
MIN_HOLDING_PERIODS = 4            # Минимальное время удержания позиции (4 периода = 20 минут на 5м)

# Настройки API Binance
API_KEY = 'ufPWU5B8qbDufrWGOKnAxmYVMOTY2BBYSGJDtHwEHHmeD7KDwH4CSzC7xIKsPc5a'
API_SECRET = 'CFGkryA2ryv8Cf7LmA8MXTPntkA7dy1FB5caTp6GrhFoWIIZCnNf6kP3JWTEuveS'

# Параметры API
API_DELAY = 1.0                    # Задержка между запросами к API (секунды)
KLINES_LIMIT = 1000                # Максимум свечек за один запрос

# =============================================================================

class PairsTradingAnalyzer:
    def __init__(self):
        """
        Инициализация анализатора парного трейдинга
        """
        # Клиент Binance с API ключами
        self.client = Client(API_KEY, API_SECRET)
        
        # Использование настроек из конфигурации
        self.symbols = SYMBOLS
        self.data = {}
        self.prices_df = None
        self.returns_df = None
        
    def fetch_futures_data(self, symbol, start_date=START_DATE, end_date=END_DATE, interval=TIMEFRAME):
        """
        Получение данных фьючерсов с Binance с учетом лимитов API
        """
        try:
            print(f"Получение данных для {symbol}...")
            
            # Преобразование дат в миллисекунды
            start_ts = int(datetime.strptime(start_date, '%Y-%m-%d').timestamp() * 1000)
            end_ts = int(datetime.strptime(end_date, '%Y-%m-%d').timestamp() * 1000)
            
            all_klines = []
            current_start = start_ts
            
            # Получение данных по частям с учетом лимита в 1000 свечек
            while current_start < end_ts:
                print(f"  Загрузка данных с {datetime.fromtimestamp(current_start/1000)}")
                
                # Получение данных кандлестиков
                klines = self.client.futures_historical_klines(
                    symbol=symbol,
                    interval=interval,
                    start_str=current_start,
                    end_str=end_ts,
                    limit=KLINES_LIMIT
                )
                
                if not klines:
                    break
                
                all_klines.extend(klines)
                
                # Обновляем начальную точку для следующего запроса
                # Берем время последней свечки + 1 интервал
                last_timestamp = int(klines[-1][0])
                
                # Рассчитываем интервал в миллисекундах
                interval_ms = self._get_interval_ms(interval)
                current_start = last_timestamp + interval_ms
                
                # Пауза между запросами
                time.sleep(API_DELAY)
                
                # Проверяем, не достигли ли мы конечной даты
                if current_start >= end_ts:
                    break
            
            if not all_klines:
                print(f"Нет данных для {symbol}")
                return None
            
            # Преобразование в DataFrame
            df = pd.DataFrame(all_klines, columns=[
                'timestamp', 'open', 'high', 'low', 'close', 'volume',
                'close_time', 'quote_asset_volume', 'number_of_trades',
                'taker_buy_base_asset_volume', 'taker_buy_quote_asset_volume', 'ignore'
            ])
            
            # Обработка данных
            df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
            df.set_index('timestamp', inplace=True)
            
            # Конвертация в числовые значения
            numeric_columns = ['open', 'high', 'low', 'close', 'volume']
            df[numeric_columns] = df[numeric_columns].astype(float)
            
            # Удаление дубликатов по индексу (времени)
            df = df[~df.index.duplicated(keep='first')]
            
            # Сортировка по времени
            df = df.sort_index()
            
            print(f"  Загружено {len(df)} свечек для {symbol}")
            print(f"  Период: {df.index[0]} - {df.index[-1]}")
            
            return df[['close']]
            
        except Exception as e:
            print(f"Ошибка при получении данных для {symbol}: {e}")
            return None
    
    def _get_interval_ms(self, interval):
        """
        Конвертация интервала в миллисекунды
        """
        interval_map = {
            '1m': 60 * 1000,
            '3m': 3 * 60 * 1000,
            '5m': 5 * 60 * 1000,
            '15m': 15 * 60 * 1000,
            '30m': 30 * 60 * 1000,
            '1h': 60 * 60 * 1000,
            '2h': 2 * 60 * 60 * 1000,
            '4h': 4 * 60 * 60 * 1000,
            '6h': 6 * 60 * 60 * 1000,
            '8h': 8 * 60 * 60 * 1000,
            '12h': 12 * 60 * 60 * 1000,
            '1d': 24 * 60 * 60 * 1000,
            '3d': 3 * 24 * 60 * 60 * 1000,
            '1w': 7 * 24 * 60 * 60 * 1000,
            '1M': 30 * 24 * 60 * 60 * 1000
        }
        return interval_map.get(interval, 15 * 60 * 1000)  # По умолчанию 15 минут
    
    def load_all_data(self, start_date=START_DATE, end_date=END_DATE):
        """
        Загрузка данных для всех символов
        """
        print("Начинаем загрузку данных...")
        print(f"Период загрузки: {start_date} - {end_date}")
        print(f"Таймфрейм: {TIMEFRAME}")
        print(f"Количество символов: {len(self.symbols)}")
        
        successful_downloads = 0
        
        for i, symbol in enumerate(self.symbols, 1):
            print(f"\n[{i}/{len(self.symbols)}] Обработка {symbol}")
            df = self.fetch_futures_data(symbol, start_date, end_date)
            if df is not None:
                self.data[symbol] = df
                successful_downloads += 1
            else:
                print(f"Не удалось загрузить данные для {symbol}")
            
            # Прогресс
            print(f"Прогресс: {i}/{len(self.symbols)} ({i/len(self.symbols)*100:.1f}%)")
            
        print(f"\nУспешно загружены данные для {successful_downloads} из {len(self.symbols)} символов")
        
        # Создание общего DataFrame с ценами закрытия
        if self.data:
            self.prices_df = pd.DataFrame({
                symbol: data['close'] for symbol, data in self.data.items()
            })
            
            # Удаление строк с NaN
            self.prices_df.dropna(inplace=True)
            
            # Расчет доходностей
            self.returns_df = self.prices_df.pct_change().dropna()
            
            print(f"Данные загружены для {len(self.data)} активов")
            print(f"Период: {self.prices_df.index[0]} - {self.prices_df.index[-1]}")
            print(f"Количество наблюдений: {len(self.prices_df)}")
        else:
            print("Не удалось загрузить данные")
    
    def calculate_correlation_analysis(self):
        """
        1. Анализ корреляции между парами активов
        """
        print("\n=== 1. АНАЛИЗ КОРРЕЛЯЦИИ ===")
        
        # 1.1 Корреляция Пирсона
        correlation_matrix = self.prices_df.corr(method='pearson')
        
        # Создание тепловой карты
        plt.figure(figsize=(16, 14))
        mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
        sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt='.3f', 
                   cmap='RdYlBu_r', center=0, square=True, cbar_kws={"shrink": .8})
        plt.title('Корреляционная матрица активов (Пирсон)', fontsize=16)
        plt.tight_layout()
        plt.show()
        
        # Топ-20 пар с наибольшей корреляцией
        pairs_correlation = []
        symbols = list(correlation_matrix.columns)
        
        for i in range(len(symbols)):
            for j in range(i+1, len(symbols)):
                pair = (symbols[i], symbols[j])
                corr_value = correlation_matrix.iloc[i, j]
                pairs_correlation.append({
                    'Pair': f"{pair[0]} - {pair[1]}",
                    'Asset1': pair[0],
                    'Asset2': pair[1],
                    'Correlation': corr_value
                })
        
        pairs_df = pd.DataFrame(pairs_correlation)
        top_20_pairs = pairs_df.nlargest(TOP_PAIRS_COUNT, 'Correlation')
        
        print(f"\nТоп-{TOP_PAIRS_COUNT} пар с наибольшей корреляцией:")
        print(top_20_pairs.to_string(index=False))
        
        # Расчет стабильности корреляции для топ пар
        correlation_stability = []
        
        for _, row in top_20_pairs.iterrows():
            asset1, asset2 = row['Asset1'], row['Asset2']
            
            # Скользящая корреляция
            rolling_corr = self.prices_df[asset1].rolling(ROLLING_WINDOW).corr(self.prices_df[asset2])
            rolling_corr = rolling_corr.dropna()
            
            if len(rolling_corr) > 0:
                corr_std = rolling_corr.std()
                corr_mean = rolling_corr.mean()
                stability_score = corr_mean / corr_std if corr_std != 0 else 0
                
                correlation_stability.append({
                    'Pair': row['Pair'],
                    'Asset1': asset1,
                    'Asset2': asset2,
                    'Static_Correlation': row['Correlation'],
                    'Rolling_Corr_Mean': corr_mean,
                    'Rolling_Corr_Std': corr_std,
                    'Stability_Score': stability_score
                })
        
        stability_df = pd.DataFrame(correlation_stability)
        
        return top_20_pairs, stability_df
    
    def cointegration_tests(self, pairs_df):
        """
        2. Тесты на коинтеграцию
        """
        print("\n=== 2. ТЕСТЫ НА КОИНТЕГРАЦИЮ ===")
        
        # Тест Энгла-Грейнджера
        print("\nТест Энгла-Грейнджера")
        
        engle_granger_results = []
        
        for _, row in pairs_df.iterrows():
            asset1, asset2 = row['Asset1'], row['Asset2']
            
            try:
                # Тест коинтеграции
                score, p_value, _ = coint(self.prices_df[asset1], self.prices_df[asset2])
                
                is_cointegrated = p_value < COINTEGRATION_P_VALUE
                
                engle_granger_results.append({
                    'Pair': row['Pair'],
                    'Asset1': asset1,
                    'Asset2': asset2,
                    'EG_Score': score,
                    'EG_P_Value': p_value,
                    'Is_Cointegrated_EG': is_cointegrated
                })
                
            except Exception as e:
                print(f"Ошибка в тесте EG для {row['Pair']}: {e}")
        
        eg_df = pd.DataFrame(engle_granger_results)
        cointegrated_eg = eg_df[eg_df['Is_Cointegrated_EG'] == True]
        
        print(f"\nПары с коинтеграцией по тесту Энгла-Грейнджера ({len(cointegrated_eg)} из {len(eg_df)}):")
        if len(cointegrated_eg) > 0:
            print(cointegrated_eg[['Pair', 'EG_Score', 'EG_P_Value']].to_string(index=False))
        else:
            print("Коинтегрированных пар не найдено")
        
        return eg_df
    
    def filter_qualified_pairs(self, top_20_pairs, stability_df, eg_df):
        """
        Фильтрация пар по всем критериям
        """
        print("\n=== ФИЛЬТРАЦИЯ ПАР ПО КРИТЕРИЯМ ===")
        
        # Критерии отбора из настроек:
        # 1. Корреляция > MIN_CORRELATION
        # 2. Входит в топ пар по корреляции
        # 3. Коинтегрирована по тесту EG
        
        high_corr_pairs = top_20_pairs[top_20_pairs['Correlation'] > MIN_CORRELATION]
        
        # Объединение всех результатов
        qualified_pairs = high_corr_pairs.copy()
        
        # Добавление результатов стабильности
        qualified_pairs = qualified_pairs.merge(
            stability_df[['Pair', 'Static_Correlation', 'Rolling_Corr_Mean', 'Rolling_Corr_Std', 'Stability_Score']], 
            on='Pair', how='left'
        )
        
        # Добавление результатов EG теста
        qualified_pairs = qualified_pairs.merge(
            eg_df[['Pair', 'EG_Score', 'EG_P_Value', 'Is_Cointegrated_EG']], 
            on='Pair', how='left'
        )
        
        # Финальная фильтрация
        final_pairs = qualified_pairs[
            (qualified_pairs['Correlation'] > MIN_CORRELATION) &
            (qualified_pairs['Is_Cointegrated_EG'] == True)
        ]
        
        print(f"\nПары, прошедшие все критерии отбора: {len(final_pairs)}")
        if len(final_pairs) > 0:
            print("Критерии отбора:")
            print(f"- Корреляция > {MIN_CORRELATION}")
            print(f"- Коинтеграция по тесту Энгла-Грейнджера (p-value < {COINTEGRATION_P_VALUE})")
            print(f"- Входят в топ-{TOP_PAIRS_COUNT} пар по корреляции")
            print("\nОтобранные пары:")
            print(final_pairs[['Pair', 'Asset1', 'Asset2']].to_string(index=False))
        else:
            print("Нет пар, удовлетворяющих всем критериям")
            print(f"Используем пары с корреляцией > {MIN_CORRELATION} для демонстрации методологии:")
            print("Критерии для демонстрации:")
            print(f"- Корреляция > {MIN_CORRELATION}")
            print(f"- Входят в топ-{TOP_PAIRS_COUNT} пар по корреляции")
            final_pairs = high_corr_pairs.head(5)  # Берем топ-5 для демонстрации
            print("\nПары для демонстрации:")
            print(final_pairs[['Pair', 'Asset1', 'Asset2']].to_string(index=False))
        
        return final_pairs
    
    def calculate_rolling_beta(self, asset1_prices, asset2_prices, lookback_period=LOOKBACK_PERIOD):
        """
        Расчет скользящей беты с использованием линейной регрессии
        """
        betas = []
        alphas = []
        
        for i in range(lookback_period, len(asset1_prices)):
            # Берем окно данных
            y = asset1_prices.iloc[i-lookback_period:i].values
            X = asset2_prices.iloc[i-lookback_period:i].values.reshape(-1, 1)
            
            # Линейная регрессия
            reg = LinearRegression().fit(X, y)
            betas.append(reg.coef_[0])
            alphas.append(reg.intercept_)
        
        # Создаем Series с правильным индексом
        beta_series = pd.Series(betas, index=asset1_prices.index[lookback_period:])
        alpha_series = pd.Series(alphas, index=asset1_prices.index[lookback_period:])
        
        return beta_series, alpha_series
    
    def spread_analysis(self, qualified_pairs):
        """
        3. Анализ спреда для отобранных пар с использованием скользящей беты
        """
        print("\n=== 3. АНАЛИЗ СПРЕДА ===")
        print(f"Используется скользящая бета с периодом lookback: {LOOKBACK_PERIOD}")
        print(f"Минимальное время удержания позиции: {MIN_HOLDING_PERIODS} периодов ({MIN_HOLDING_PERIODS * 5} минут)")
        
        spread_results = []
        
        for _, row in qualified_pairs.iterrows():
            asset1, asset2 = row['Asset1'], row['Asset2']
            pair_name = row['Pair']
            
            print(f"\nАнализ пары: {pair_name}")
            
            # Выводим все доступные метрики для пары
            print(f"Static Correlation: {row.get('Static_Correlation', row['Correlation']):.6f}")
            print(f"Rolling Correlation Mean: {row.get('Rolling_Corr_Mean', 'N/A')}")
            print(f"Rolling Correlation Std: {row.get('Rolling_Corr_Std', 'N/A')}")
            print(f"EG Score: {row.get('EG_Score', 'N/A')}")
            print(f"EG P-Value: {row.get('EG_P_Value', 'N/A')}")
            
            # 3.1 Расчет скользящей беты
            asset1_prices = self.prices_df[asset1]
            asset2_prices = self.prices_df[asset2]
            
            rolling_beta, rolling_alpha = self.calculate_rolling_beta(asset1_prices, asset2_prices, LOOKBACK_PERIOD)
            
            print(f"Скользящая бета - среднее: {rolling_beta.mean():.6f}, std: {rolling_beta.std():.6f}")
            print(f"Скользящая альфа - среднее: {rolling_alpha.mean():.6f}, std: {rolling_alpha.std():.6f}")
            
            # 3.2 Расчет спреда со скользящей бетой
            # Выравниваем индексы
            common_index = rolling_beta.index
            spread = asset1_prices.loc[common_index] - rolling_beta * asset2_prices.loc[common_index]
            
            # 3.3 Расчет скользящего Z-Score
            spread_mean = spread.rolling(LOOKBACK_PERIOD).mean()
            spread_std = spread.rolling(LOOKBACK_PERIOD).std()
            spread_zscore = (spread - spread_mean) / spread_std
            
            # Удаляем NaN значения
            spread_zscore = spread_zscore.dropna()
            spread = spread.loc[spread_zscore.index]
            
            # 3.4 Тест на стационарность (ADF) для спреда
            adf_result = adfuller(spread.dropna())
            adf_statistic = adf_result[0]
            adf_p_value = adf_result[1]
            is_stationary = adf_p_value < ADF_P_VALUE
            
            print(f"ADF статистика: {adf_statistic:.6f}")
            print(f"ADF p-value: {adf_p_value:.6f}")
            print(f"Спред стационарен: {is_stationary}")
            
            # Сохранение результатов
            spread_results.append({
                'Pair': pair_name,
                'Asset1': asset1,
                'Asset2': asset2,
                'Rolling_Beta_Mean': rolling_beta.mean(),
                'Rolling_Beta_Std': rolling_beta.std(),
                'Rolling_Alpha_Mean': rolling_alpha.mean(),
                'Rolling_Alpha_Std': rolling_alpha.std(),
                'ADF_Statistic': adf_statistic,
                'ADF_P_Value': adf_p_value,
                'Is_Stationary': is_stationary,
                'Static_Correlation': row.get('Static_Correlation', row['Correlation']),
                'Rolling_Corr_Mean': row.get('Rolling_Corr_Mean', None),
                'Rolling_Corr_Std': row.get('Rolling_Corr_Std', None),
                'EG_Score': row.get('EG_Score', None),
                'EG_P_Value': row.get('EG_P_Value', None),
                'Spread': spread,
                'Spread_ZScore': spread_zscore,
                'Rolling_Beta': rolling_beta,
                'Rolling_Alpha': rolling_alpha
            })
            
            # Визуализация
            self.plot_spread_analysis_with_rolling_beta(
                spread, spread_zscore, rolling_beta,
                asset1_prices, asset2_prices,
                pair_name, asset1, asset2
            )
        
        return spread_results
    
    def plot_spread_analysis_with_rolling_beta(self, spread, spread_zscore, rolling_beta,
                                              asset1_prices, asset2_prices,
                                              pair_name, asset1, asset2):
        """
        Визуализация анализа спреда с учетом скользящей беты и минимального времени удержания
        """
        fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
        
        # 1. График двух активов с сигналами входа/выхода
        # Нормализация цен для визуального сравнения
        common_index = spread_zscore.index
        price1_norm = (asset1_prices.loc[common_index] / asset1_prices.loc[common_index].iloc[0]) * 100
        price2_norm = (asset2_prices.loc[common_index] / asset2_prices.loc[common_index].iloc[0]) * 100
        
        ax1.plot(price1_norm.index, price1_norm, label=asset1, linewidth=1.5, alpha=0.8)
        ax1.plot(price2_norm.index, price2_norm, label=asset2, linewidth=1.5, alpha=0.8)
        
        # Генерация сигналов на основе Z-Score с учетом минимального времени удержания
        entry_threshold = ENTRY_THRESHOLD
        exit_threshold = EXIT_THRESHOLD
        min_holding = MIN_HOLDING_PERIODS
        
        # Поиск точек входа и выхода
        position = 0  # 0 - нет позиции, 1 - лонг спред, -1 - шорт спред
        entry_points_long = []
        entry_points_short = []
        exit_points = []
        periods_in_position = 0
        
        for i in range(len(spread_zscore)):
            current_zscore = spread_zscore.iloc[i]
            
            if position == 0:  # Нет позиции
                if current_zscore > entry_threshold:  # Спред высокий - шорт спред
                    position = -1
                    periods_in_position = 0
                    entry_points_short.append((spread_zscore.index[i], price1_norm.iloc[i], price2_norm.iloc[i]))
                elif current_zscore < -entry_threshold:  # Спред низкий - лонг спред
                    position = 1
                    periods_in_position = 0
                    entry_points_long.append((spread_zscore.index[i], price1_norm.iloc[i], price2_norm.iloc[i]))
            
            elif position != 0:  # Есть позиция
                periods_in_position += 1
                # Выход только после минимального времени удержания
                if periods_in_position >= min_holding and abs(current_zscore) < exit_threshold:
                    exit_points.append((spread_zscore.index[i], price1_norm.iloc[i], price2_norm.iloc[i]))
                    position = 0
                    periods_in_position = 0
        
        # Отображение сигналов на графике цен
        for entry in entry_points_long:
            ax1.scatter(entry[0], entry[1], color='green', marker='^', s=100, alpha=0.8, 
                       label='Long Asset1' if entry == entry_points_long[0] else "")
            ax1.scatter(entry[0], entry[2], color='red', marker='v', s=100, alpha=0.8, 
                       label='Short Asset2' if entry == entry_points_long[0] else "")
        
        for entry in entry_points_short:
            ax1.scatter(entry[0], entry[1], color='red', marker='v', s=100, alpha=0.8, 
                       label='Short Asset1' if entry == entry_points_short[0] else "")
            ax1.scatter(entry[0], entry[2], color='green', marker='^', s=100, alpha=0.8, 
                       label='Long Asset2' if entry == entry_points_short[0] else "")
        
        for exit_point in exit_points:
            ax1.scatter(exit_point[0], exit_point[1], color='yellow', marker='o', s=80, alpha=0.8, 
                       edgecolors='black', label='Exit' if exit_point == exit_points[0] else "")
            ax1.scatter(exit_point[0], exit_point[2], color='yellow', marker='o', s=80, alpha=0.8, 
                       edgecolors='black')
        
        ax1.set_title(f'Цены активов с сигналами входа/выхода: {pair_name}')
        ax1.set_ylabel('Нормализованная цена')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 2. Скользящая бета
        ax2.plot(rolling_beta.index, rolling_beta, linewidth=1.5, color='purple', label='Скользящая бета')
        ax2.axhline(y=rolling_beta.mean(), color='purple', linestyle='--', alpha=0.5, 
                   label=f'Среднее: {rolling_beta.mean():.4f}')
        ax2.set_title(f'Скользящая бета (lookback={LOOKBACK_PERIOD})')
        ax2.set_ylabel('Beta')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # 3. Z-Score с уровнями входа/выхода
        ax3.plot(spread_zscore.index, spread_zscore, linewidth=1, alpha=0.7, color='blue')
        ax3.axhline(y=0, color='black', linestyle='-', alpha=0.5)
        ax3.axhline(y=entry_threshold, color='red', linestyle='--', alpha=0.7, 
                   label=f'Вход ±{entry_threshold}σ')
        ax3.axhline(y=-entry_threshold, color='red', linestyle='--', alpha=0.7)
        ax3.axhline(y=exit_threshold, color='orange', linestyle=':', alpha=0.7, 
                   label=f'Выход ±{exit_threshold}σ')
        ax3.axhline(y=-exit_threshold, color='orange', linestyle=':', alpha=0.7)
        
        # Отображение сигналов на Z-Score
        for entry in entry_points_long + entry_points_short:
            idx = spread_zscore.index.get_loc(entry[0])
            ax3.scatter(entry[0], spread_zscore.iloc[idx], 
                       color='green' if entry in entry_points_long else 'red', 
                       marker='^' if entry in entry_points_long else 'v', s=100, alpha=0.8)
        
        for exit_point in exit_points:
            idx = spread_zscore.index.get_loc(exit_point[0])
            ax3.scatter(exit_point[0], spread_zscore.iloc[idx], color='yellow', 
                       marker='o', s=80, alpha=0.8, edgecolors='black')
        
        ax3.set_title(f'Z-Score спреда (lookback={LOOKBACK_PERIOD}, min hold={MIN_HOLDING_PERIODS})')
        ax3.set_ylabel('Z-Score')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        # Вывод статистики по сигналам
        print(f"\nСтатистика сигналов для {pair_name}:")
        print(f"Количество сделок лонг спреда: {len(entry_points_long)}")
        print(f"Количество сделок шорт спреда: {len(entry_points_short)}")
        print(f"Количество закрытых сделок: {len(exit_points)}")
        print(f"Минимальное время удержания: {MIN_HOLDING_PERIODS} периодов ({MIN_HOLDING_PERIODS * 5} минут)")
    
    def run_full_analysis(self):
        """
        Запуск полного анализа
        """
        print("КОЛИЧЕСТВЕННАЯ АНАЛИТИКА ПАРНОГО ТРЕЙДИНГА")
        print("=" * 50)
        
        # Загрузка данных
        self.load_all_data()
        
        if self.prices_df is None or len(self.prices_df) == 0:
            print("Ошибка: Не удалось загрузить данные")
            return
        
        # 1. Анализ корреляции
        top_20_pairs, stability_df = self.calculate_correlation_analysis()
        
        # 2. Тесты на коинтеграцию
        eg_df = self.cointegration_tests(top_20_pairs)
        
        # 3. Фильтрация пар по критериям
        qualified_pairs = self.filter_qualified_pairs(top_20_pairs, stability_df, eg_df)
        
        # 4. Анализ спреда
        if len(qualified_pairs) > 0:
            spread_results = self.spread_analysis(qualified_pairs)
        else:
            print("Нет пар, подходящих для парного трейдинга по заданным критериям")


# Запуск анализа
if __name__ == "__main__":
    analyzer = PairsTradingAnalyzer()
    analyzer.run_full_analysis()
```
